```python
# Define the transition probability matrix of the Markov chain
P = np.array([[0.2, 0.8, 0.0],
              [0.5, 0.0, 0.5],
              [0.3, 0.3, 0.4]])

# Find the dominant eigenvalue and eigenvector using power iteration
def power_iteration(A, max_iter=1000, tol=1e-6):
    n = A.shape[0]
    x = np.ones(n) / n
    for i in range(max_iter):
        x_new = A @ x
        if np.linalg.norm(x_new - x) < tol:
            break
        x = x_new
    lambda_max = x_new @ A @ x_new / (x_new @ x_new)
    return lambda_max, x_new / np.sum(x_new)

lambda_max, pi = power_iteration(P)

# Calculate the rate of convergence and convergence time
rate_of_conv = 1 - lambda_max
conv_time = -1 / np.log(lambda_max)

print("Rate of convergence:", rate_of_conv)
print("Convergence time:", conv_time)
```
